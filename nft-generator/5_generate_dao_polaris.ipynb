```python
from PIL import Image 
from IPython.display import display 
import random
import json
```

```python
# Each image is made up a series of traits
# The weightings for each trait drive the rarity and add up to 100%


background = [
    "DAO Polaris",
]
accessories_eyewear = [
    "No Eyewear",
    "Silver LED Eyewear",
]
hair_front = [
    "Bald",
    "Blue Curl Hair",
    "Silver Circle Hair",
]
accessories_earring = [
    "No Earring",
    "Blue ATEM Earring",
]
accessories_cuff = [
    "No Cuff",
    "Silver Star Nose Cuff",
]
accessories_necklace = [
    "No Necklace",
]
eyebrow = [
    "No Eyebrow",
    "Purple Fire Eyelash",
]
eye_left = [
    "Desert Eyes",
    "Ocean Eyes",
    "Voyager  Eyes",
    "Microgravity Eyes",
]
eye_right = [
    "Desert Eyes",
    "Ocean Eyes",
    "Voyager  Eyes",
    "Microgravity Eyes",
]
eye_cloth = [
    "White Rock Leather Jacket",
    "Green Fur Hoodie",
]
eye_skin = [
    "Polaris Skin",
]
hair_back = [
    "Bald",
    "Blue Curl Hair",
    "Silver Circle Hair",
]

# Dictionary variable for each trait. 
# Eech trait corresponds to its file name


background_files = {
    "DAO Polaris": "B_05",
}
accessories_eyewear_files = {
    "No Eyewear": "AE_00",
    "Silver LED Eyewear": "AE_02_2",
}
hair_front_files = {
    "Bald" :"HF_00",
    "Blue Curl Hair": "HF_02_2",
    "Silver Circle Hair": "HF_07_1",
}
accessories_earring_files = {
    "No Earring": "AR_00",
    "Blue ATEM Earring": "AR_01_2",
}
accessories_cuff_files = {
    "No Cuff": "AC_00",
    "Silver Star Nose Cuff": "AC_02_1",
}
accessories_necklace_files = {
    "No Necklace": "AN_00",
}
eyebrow_files = {
    "No Eyebrow": "EB_00",
    "Purple Fire Eyelash": "EB_05_1",
}
eye_left_files = {
    "Desert Eyes": "EL_01",
    "Ocean Eyes": "EL_03",
    "Voyager  Eyes": "EL_10",
    "Microgravity Eyes": "EL_12",
}
eye_right_files = {
    "Desert Eyes": "ER_01",
    "Ocean Eyes": "ER_03",
    "Voyager  Eyes": "ER_10",
    "Microgravity Eyes": "ER_12",
}
eye_cloth_files = {
    "White Rock Leather Jacket": "C_01_2",
    "Green Fur Hoodie": "C_03_3",
}
eye_skin_files = {
    "Polaris Skin": "S_09",
}
hair_back_files = {
    "Bald": "HB_00",
    "Blue Curl Hair": "HB_02_2",
    "Silver Circle Hair": "HB_07_1",
}
```

```python
## Generate Traits

TOTAL_IMAGES = 375 # Number of random unique images we want to generate

all_images = [] 
HF_08_ITEMS = [
    "Silver A Zone Helmet",
    "Brown Patterned A Zone Helmet",
    "Purple A Zone Helmet",
    "Purple Patterned A Zone Helmet"
]
# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #

    # For each trait category, select a random trait based on the weightings 
    new_image ["Background"] = random.choices(background)[0]
    new_image ["accessories_eyewear"] = random.choices(accessories_eyewear)[0]
    new_image ["hair_front"] = random.choices(hair_front)[0]
    new_image ["accessories_earring"] = random.choices(accessories_earring)[0]
    new_image ["accessories_cuff"] = random.choices(accessories_cuff)[0]
    new_image ["accessories_necklace"] = random.choices(accessories_necklace)[0]
    new_image ["eyebrow"] = random.choices(eyebrow)[0]
    new_image ["eye_left"] = random.choices(eye_left)[0]
    new_image ["eye_right"] = new_image ["eye_left"]
    new_image ["eye_cloth"] = random.choices(eye_cloth)[0]
    new_image ["eye_skin"] = random.choices(eye_skin)[0]
    new_image ["hair_back"] = new_image ["hair_front"]

    if new_image ["hair_front"] in HF_08_ITEMS:
        new_image ["accessories_eyewear"] = "AE_00"
    
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
```

```python
#### Generate Metadata for each Image    

f = open('./metadata/all-traits.json',) 
data = json.load(f)


IMAGES_BASE_URI = "ADD_IMAGES_BASE_URI_HERE"
PROJECT_NAME = "Atem NFT Passport - DAO Polaris"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("Background", i["Background"]))
    token["attributes"].append(getAttribute("accessories_eyewear", i["accessories_eyewear"]))
    token["attributes"].append(getAttribute("hair_front", i["hair_front"]))
    token["attributes"].append(getAttribute("accessories_earring", i["accessories_earring"]))
    token["attributes"].append(getAttribute("accessories_cuff", i["accessories_cuff"]))
    token["attributes"].append(getAttribute("accessories_necklace", i["accessories_necklace"]))
    token["attributes"].append(getAttribute("eyebrow", i["eyebrow"]))
    token["attributes"].append(getAttribute("eye_left", i["eye_left"]))
    token["attributes"].append(getAttribute("eye_right", i["eye_right"]))
    token["attributes"].append(getAttribute("eye_cloth", i["eye_cloth"]))
    token["attributes"].append(getAttribute("eye_skin", i["eye_skin"]))
    token["attributes"].append(getAttribute("hair_back", i["hair_back"]))

    with open('./metadata/' + str(token_id), 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
